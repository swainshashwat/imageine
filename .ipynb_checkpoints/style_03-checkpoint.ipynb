```python
import numpy as np
import keras.backend as K
from scipy.misc import imsave, imresize
from scipy.optimize import fmin_l_bfgs_b
from keras.applications import vgg19
from keras.preprocessing.image import img_to_array, load_img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dims = {
    'height': 512,
    'width': 690,
    'channels': 3
}

style_path = 'data/melodyonight.jpg'
content_path = 'data/megan_fox.jpg'
```

```python
def preprocess_img(img_path, dim=dims):
    """
    Preprocesses image input to a
    predefined matrix for
    dimension matching for the model
    """
    # read img
    img = load_img(img_path)
    # to array
    img = img_to_array(img)
    
    # resize
    img = imresize(img, (dim['height'], dim['width'], 3))
    
    return img
```

```python
def preprocess_img_for_network(img):
    img = preprocess_img(img)
    img = img.astype('float64')
    img = np.expand_dims(img, axis=0)
    
    img = vgg19.preprocess_input(img)
    
    return img
```

```python
def display_image(img, title=None):
    """
    Preprocess image for display
    """
    plt.figure()
    
    plt.imshow(img)
    
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
content_img = preprocess_img(content_path)
style_img = preprocess_img(style_path)
```

```python
display_image(content_img, title='Content Image')
display_image(style_img, title='Style Image')
```

```python
# setting image variable and placeholder for network input
content_img_0 = K.variable(preprocess_img_for_network(
                            content_path))
style_img_0 = K.variable(preprocess_img_for_network(
                            style_path))

gen_img_0 = K.placeholder(shape=(
                1, dims['height'], dims['width'], 3))
```

```python
# creating the input tensor
input_tensor = K.concatenate([content_img_0,
                              style_img_0,
                              gen_img_0], axis=0)
```

```python
# building the mode
model = vgg19.VGG19(include_top=False,
                    weights='imagenet',
                   input_tensor=input_tensor)
```

```python
def content_loss(gen, content):
    """
    Eculidian distance Loss between
    Content image and Generated image
    """
    return K.sum(K.square(gen - content))
```
